```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="hi i am learning the langchain for agentic ai and ai agents its really amazing"
query_result=embeddings.embed_query(text)
print(query_result)
```

### document ingestion -> splitter -> Embedding -> Vector store 

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("sample3.txt")
document=loader.load()
document
```

```python
# now lets do text splitting 
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
split_doc=text_splitter.split_documents(document)
```

```python
split_doc[0].page_content
```

```python
# now lets do embeddings + vector store chroma database
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(split_doc,embeddings_1024)
db
```

```python
# Retrieve the results from query vectorstore db with similarity search
query="Quem si tenueris, non modo meum Ciceronem, sed etiam me ipsum abducas licebit."

retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
